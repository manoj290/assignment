```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
from numpy.random import randn
```

```python
np.random.seed(4)
matrix_data=randn(10,4)
heading="A B C D".split()
df=pd.DataFrame(data=matrix_data,columns=heading)
df
```

```python
df.plot.bar()
```

```python
df.plot.bar()
```

```python
df.plot.barh()
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.tail()
```

```python
iris.sepal_length.plot(kind='hist',bins=35)
```

```python
df=iris.drop(['species'],axis=1)
```

```python
df.tail()
```

```python
df.plot(kind='bar',figsize=(100,10))
```

```python
df.plot(kind='hist',color='r',alpha=0.4,figsize=(10,5))
plt.show()
```

```python
df.hist(color='r',alpha=0.5,figsize=(10,5))
plt.show()
```

#### BoxPlot

```python
df.plot(kind="box",color='r',figsize=(10,5))
```

```python
df.plot(kind='box',figsize=(10,5),vert=False)
```

#### Scatter_Matrix

```python
from pandas.plotting import scatter_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
#kde=kernal density estimation
scatter_matrix(df,diagonal="kde",figsize=(8,8))
plt.show()
```

#### Pie Chart

```python
d=df.iloc[0]
d
```

```python
d.plot.pie(figsize=(5,5))
plt.show()
```

```python
d=df.head(3).T
d
```

```python
d.plot.pie(figsize=(25,20),subplots=True)
plt.show()
```

```python
d.plot.pie(subplots=True,figsize=(60,50),autopct='%0.2f',fontsize=50)
plt.show()
```

#### Scatter Plot

```python
df.plot.scatter(x='petal_length',y='sepal_length')
plt.show()
```

```python
df.plot(kind='area')
```

```python
ax=df.plot(kind='scatter',x='sepal_length',y='petal_length',label='Length',color='r')
df.plot.scatter(x='sepal_width',y='petal_width',ax=ax,label='Width',color='y')
plt.show()
```
