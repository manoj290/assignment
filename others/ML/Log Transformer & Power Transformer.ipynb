```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/training-ml/Files/main/diabetes.csv")
data
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
data.Glucose=data.Glucose.replace(0,data.Glucose.mean())
data.BloodPressure=data.BloodPressure.replace(0,data.BloodPressure.mean())
data.SkinThickness=data.SkinThickness.replace(0,data.SkinThickness.mean())
data.Insulin=data.Insulin.replace(0,data.Insulin.mean())
data.BMI=data.BMI.replace(0,data.BMI.mean())
```

```python
data.describe()
```

```python
plt.figure(figsize=(20,25))
graph=1
for col in data:
    if graph<12:
        ax=plt.subplot(4,3,graph)
        sns.distplot(data[col])
        plt.xlabel(col)
    graph+=1
plt.tight_layout()
```

## Log Transformation

###### In this we take log of th column & it convert skewed data into normal distribution(less skewed data). And it can be applied only to the continous variables.

```python
data.BloodPressure=np.log(data.BloodPressure)
data.Glucose=np.log(data.Glucose)
data.SkinThickness=np.log(data.SkinThickness)
data.Insulin=np.log(data.Insulin)
data.BMI=np.log(data.BMI)
data.DiabetesPedigreeFunction=np.log(data.DiabetesPedigreeFunction)
```

```python
plt.figure(figsize=(20,25))
graph=1
for col in data:
    if graph<12:
        ax=plt.subplot(4,3,graph)
        sns.distplot(data[col])
        plt.xlabel(col)
    graph+=1
plt.tight_layout()
```

## Power Transformation

###### It is a technique to transform data. In this it stabilize the data & is transformed to normal disribution like.

#### Box-Cox : it is applied only the values greater than 0.

#### Yeo-Johnson: It is applied to both positive & negative values.

```python
data=pd.read_csv("https://raw.githubusercontent.com/training-ml/Files/main/diabetes.csv")
data
```

```python
data.Glucose=data.Glucose.replace(0,data.Glucose.mean())
data.BloodPressure=data.BloodPressure.replace(0,data.BloodPressure.mean())
data.SkinThickness=data.SkinThickness.replace(0,data.SkinThickness.mean())
data.Insulin=data.Insulin.replace(0,data.Insulin.mean())
data.BMI=data.BMI.replace(0,data.BMI.mean())
```

```python
features=['Glucose','BloodPressure','SkinThickness',"Insulin","BMI","Age"]
```

```python
plt.figure(figsize=(20,25))
graph=1
for col in data[features]:
    if graph<12:
        ax=plt.subplot(4,3,graph)
        sns.distplot(data[col])
        plt.xlabel(col)
    graph+=1
plt.tight_layout()
```

```python
from sklearn.preprocessing import PowerTransformer
scalar=PowerTransformer(method='box-cox')
data[features]=scalar.fit_transform(data[features].values)
data[features]
```

```python
plt.figure(figsize=(20,25))
graph=1
for col in data[features]:
    if graph<12:
        ax=plt.subplot(4,3,graph)
        sns.distplot(data[col])
        plt.xlabel(col)
    graph+=1
plt.tight_layout()
```
