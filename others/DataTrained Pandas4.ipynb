## Operations used to handle Nan 

```python
#Creating DataFrame with missing values
raw_data={'first_name':['manoj',np.nan,'bongu','pranshu','suhaib'],
          'last_name':['saxena',np.nan,'vaish','pandey','khan'],
          'age':[24,np.nan,22,26,12],
          'sex':['m',np.nan,'f','m','m'],
          'preTest':[5,np.nan,np.nan,np.nan,4],
          'postTest':[32,np.nan,25,15,36]
          
         }
df=pd.DataFrame(raw_data)
df
```

#### Drop row with missing values

```python
#Drop missing values
dfDrop=df.dropna()
dfDrop
```

```python
dfDrop=df.dropna(how='all')
dfDrop
```

```python
dfDrop=df.dropna(how='any')
dfDrop
```

#### Drop column with missing values

```python
df['location']=np.nan
df
```

```python
df.dropna(axis=1,how='all')
```

```python
df
```

```python
df.fillna(0)
```

```python
df
```

```python
df['preTest'].fillna(df['preTest'].mean(),inplace=True)
df
```

### **filling missing values groupby sex mean values to postTest column

```python
df['postTest'].fillna(df.groupby('sex')['postTest'].transform("mean"),inplace=True)
df
```

### select some rows & ignoring missing values

```python
df[df['sex'].notnull() & df['age'].notnull()]
```

```python
df
```

```python
#back_fill
df.fillna(method='bfill')
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='ffill',axis=1)
```
