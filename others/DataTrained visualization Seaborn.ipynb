## Numerical Data Plotting
1.relplot()

2.Scatterplot()

3.LinePlot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.relplot(x='total_bill',y="tip",data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker')
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',size='size',data=df)
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',size='size',data=df)
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=df,hue='smoker',size='size')
```

#### Categorical Data Plotting
1.catplot()
2.boxplot()
3.stipplot()
4.swarmplot()


```python
sns.catplot(x='day', y='total_bill',data=df)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.catplot(x='day',y='total_bill',kind='swarm',hue='size',data=df)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=df)
```
